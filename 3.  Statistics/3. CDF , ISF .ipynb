# **Q 6**

```python
sam =  np.array ( [108.98, 152.22, 111.45, 110.59, 127.46, 107.26, 93.32, 91.97, 111.56, 75.71, 128.58, 135.11] )

sigma   =  np.std (sam, ddof=1)
```

### **T - Dist**

```python
stats.t.isf  ( 0.975,  12-1,  120,  sigma / np.sqrt(12) )
```

```python
stats.t.isf  ( 0.025,  12-1,  120,  sigma / np.sqrt(12) )
```

* Sample mean (i.e. 120) lies within above range
* Hence , no sig deviation between pop mean & sam mean
* Thus , accept null

### **T - scale**

```python
(112.85 - 120)   /   ( sigma / np.sqrt(12) )
```

```python
stats.t.isf  (0.975,  12-1,  0,  1)
```

```python
stats.t.isf  (0.025,  12-1,  0,  1)
```

* Sample mean (i.e. -1.19) lies within above range
* Hence , no sig deviation between pop mean & sam mean
* Thus , accept null

### **P value**

```python
stats.t.cdf ( 112.85  ,  12-1  ,  120  ,  sigma / np.sqrt(12) )  *  2
```

* p > alpha
* Accept Null
* Hence , no sig deviation between pop mean & sam mean
