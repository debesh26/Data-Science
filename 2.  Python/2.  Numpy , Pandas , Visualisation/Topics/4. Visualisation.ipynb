# 1. **Line Plot**

### With List

```python
a = [1,5,10,15,20]
b = [5,15,10,20,18]

plt.plot (a,b)
```

* Two Lines in 1 graph

```python
a = [1,5,10,15,20]
b = [5,15,10,20,18]
c = [4,8,5,9,6]

plt.plot (a,b)
plt.plot (a,c)
```

* Legend , Title , Label 

```python
India = [100,200,150,300]
US    = [200,300,250,350]
Month = ['Jan','Feb','Mar','Apr']

plt.plot (Month,India, label='IND')
plt.plot (Month,US,    label='US' )
plt.legend()                           # to display the labels

plt.title  ('Covid Count')
plt.xlabel ('Month')
plt.ylabel ('Count')

plt.show()
```

# 4. **Subplot**

* 1 row , 4 cols

```python
fig , pos  =  plt.subplots (1,4, figsize=(15,5))


df ['sex']    .value_counts()  .plot  ( kind='pie'  ,  ax = pos [0] )
df ['smoker'] .value_counts()  .plot  ( kind='pie'  ,  ax = pos [1] )
df ['day']    .value_counts()  .plot  ( kind='bar'  ,  ax = pos [2] )
df ['time']   .value_counts()  .plot  ( kind='bar' ,   ax = pos [3] )

plt.show()
```

* 2 rows , 2 cols

```python
fig , pos = plt.subplots (2,2, figsize=(12,7))

p = pos.ravel()

df ['sex']    .value_counts()  .plot  ( kind='pie'  ,  ax = p[0] )
df ['smoker'] .value_counts()  .plot  ( kind='pie'  ,  ax = p[1] )
df ['day']    .value_counts()  .plot  ( kind='bar'  ,  ax = p[2] )
df ['time']   .value_counts()  .plot  ( kind='bar' ,   ax = p[3] )

plt.show()
```
